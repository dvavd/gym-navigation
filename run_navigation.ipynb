```python
import gymnasium as gym
import os
from stable_baselines3 import PPO
from stable_baselines3.common.vec_env import DummyVecEnv
from stable_baselines3.common.monitor import Monitor
from stable_baselines3.common.callbacks import CheckpointCallback

log_dir = "logs"
os.makedirs(log_dir, exist_ok=True)
model_dir = "models"
os.makedirs(model_dir, exist_ok=True)

env_name = 'gym_navigation:NavigationGoal-v0'
```

```python
def make_env(env_name : str):
    env = gym.make(env_name, render_mode=None)
    env = Monitor(env, log_dir)
    return env
```

```python
env = DummyVecEnv([lambda: make_env(env_name)])

model = PPO(
    "MlpPolicy",
    env,
    verbose=1,
    learning_rate=3e-4,
    n_steps=2048,
    batch_size=64,
    n_epochs=10,
    gamma=0.99,
    tensorboard_log=log_dir
)

checkpoint_callback = CheckpointCallback(
    save_freq=10000,
    save_path=model_dir,
    name_prefix="navigation_model"
)

TIMESTEPS = 2_000_000
model.learn(
    total_timesteps=TIMESTEPS,
    callback=checkpoint_callback,
    progress_bar=True
)

model.save(f"{model_dir}/navigation_final")

env.close()
```

```python
import time

model_path = "models/navigation_final"
model = PPO.load(model_path)

env = gym.make('gym_navigation:NavigationGoal-v0', render_mode='human')

observation, info = env.reset()
total_reward = 0
episode_count = 0
max_episodes = 15 

try:
    while episode_count < max_episodes:
        action, _states = model.predict(observation, deterministic=False)
        
        observation, reward, terminated, truncated, info = env.step(action)
        
        total_reward += reward
        
        if terminated or truncated:
            print(f"Episode {episode_count+1} finished with reward: {total_reward}")
            episode_count += 1
            total_reward = 0
            observation, info = env.reset()
            
            if episode_count < max_episodes:
                print("Starting next episode in 2 seconds...")
                time.sleep(1)
                
finally:
    env.close()
```
